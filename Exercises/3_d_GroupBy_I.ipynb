### Step 3. Assign it to a variable called drinks.

```python
drinks=pd.read_csv(path+'/DATA/drinks.csv')
```

### Step 4. Which continent drinks the most beer on average?

```python
gruppiert=drinks.groupby(['continent']).mean()
print(gruppiert)
maxi=gruppiert['beer_servings'].max()
print(maxi)
gruppiert.loc[gruppiert['beer_servings'] == maxi, ['beer_servings']]
```

### Step 5. For each continent print the statistics for wine consumption.

```python
Kontinente=gruppiert.index
#print(Kontinente[0])
Afrika_w=drinks.loc[drinks['continent']==Kontinente[0],['country','wine_servings']]
print(Afrika_w['country'])

```

```python

fig, ax = plt.subplots(figsize=(50,10))
ax.bar(Afrika_w['country'], Afrika_w['wine_servings'])

ax.set_xlabel('Country')
ax.set_ylabel('wine_servigs')
```
